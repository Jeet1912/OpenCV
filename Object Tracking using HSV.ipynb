```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
def nothing(x):
    pass

cv2.namedWindow('OT using HSV')
cv2.createTrackbar('LowerH','OT using HSV',0,255,nothing)
cv2.createTrackbar('LowerS','OT using HSV',0,255,nothing)
cv2.createTrackbar('LowerV','OT using HSV',0,255,nothing)
cv2.createTrackbar('UpperH','OT using HSV',255,255,nothing)
cv2.createTrackbar('UpperS','OT using HSV',255,255,nothing)
cv2.createTrackbar('UpperV','OT using HSV',255,255,nothing)

while True:
    marbles = cv2.imread('/Users/jxxt/Downloads/marbles.jpeg')
    marbles = cv2.resize(marbles,(300,300))
    hsv = cv2.cvtColor(marbles,cv2.COLOR_BGR2HSV)

    l_h = cv2.getTrackbarPos('LowerH','OT using HSV')
    l_s = cv2.getTrackbarPos('LowerS','OT using HSV')
    l_v = cv2.getTrackbarPos('LowerV','OT using HSV')
    
    u_h = cv2.getTrackbarPos('UpperH','OT using HSV')
    u_s = cv2.getTrackbarPos('UpperS','OT using HSV')
    u_v = cv2.getTrackbarPos('UpperV','OT using HSV')
    
    lower_b = np.array([l_h,l_s,l_v])
    upper_b = np.array([u_h,u_s,u_v])
    
    mask = cv2.inRange(hsv,lower_b,upper_b)

    res = cv2.bitwise_and(marbles,marbles,mask=mask)

    cv2.imshow('marbles',marbles)
    cv2.imshow('hsv',mask)
    cv2.imshow('result',res)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
        
cv2.destroyAllWindows()
for i in range (1,5):
    cv2.waitKey(1)
```
